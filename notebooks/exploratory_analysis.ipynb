# Exploratory analysis

This notebook inspects the synthetic monitoring dataset and the reproducible
experiment outputs. All experiment results shown here are from the offline mock
provider and are simulation, not measurements of any real model.

Prerequisites: run `python -m src.data_generation` and
`python -m src.experiment_runner` from the repository root first.

```python
import sys, json
from pathlib import Path
sys.path.append(str(Path.cwd().parent))
import pandas as pd

df = pd.read_csv('../data/synthetic_monitoring_data.csv')
print('cases:', len(df))
print('anomaly prevalence:', round(df.ground_truth_anomaly.mean(), 3))
df.scenario_type.value_counts()
```

```python
# Ground-truth anomaly rate by scenario type
df.groupby('scenario_type').ground_truth_anomaly.mean().sort_values()
```

```python
# Experiment-level results (mock mode)
exp = pd.read_csv('../outputs/experiment_results.csv')
cols = ['experiment_id','prompt_version','include_deterministic_evidence',
        'reviewer_enabled','pre_precision','pre_recall','pre_f1',
        'pre_unsupported_claim_rate','pre_expected_calibration_error','baseline_f1']
exp[exp.grid=='mock_main'][cols].round(3)
```

```python
# Prompt-variant comparison
m = exp[exp.grid=='mock_main']
m.groupby('prompt_version')[['pre_false_positive_rate','pre_unsupported_claim_rate',
                             'pre_expected_calibration_error']].mean().round(3)
```
